# Ganga Data Selection 

Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

Reading dataset

```python
df = pd.read_csv('Ganga_data.csv')
```

```python
df
```

Applying Gaussian Distribution formula to check whether the data is in normal distribution or not.

```python
#cond
cond = df['COND']
cond
```

```python
cond.shape
```

```python
cond_pdf = np.zeros([46983,1])
```

```python
mu = np.mean(cond)
sigma = np.std(cond)
```

```python
for i in range(0, 46983):
    x = 1/(np.sqrt(2*np.pi)*sigma)*np.exp(-np.power((cond[i] - mu)/sigma, 2)/2)
    cond_pdf[i] = x
```

```python
cond_pdf
```

```python
plt.figure(figsize = (15,7))
plt.scatter(cond, cond_pdf, alpha = 0.5, marker = '.')
plt.xlabel('COND')
plt.ylabel('COND_PDF')
```
